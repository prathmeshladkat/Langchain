```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is test document for OPENAI EMBEDDINGS"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
##Data Ingestion
from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt")
docs = loader.load()
docs
```

```python
##Data splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20)
final_documents = text_splitter.split_documents(docs)
final_documents
```

```python
## vector embedding and Vector storesDB
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_documents, embeddings)
db
```

### Retrival 


```python
query="scalable backend systems."
retrived_results=db.similarity_search(query)
print(retrived_results)
```
