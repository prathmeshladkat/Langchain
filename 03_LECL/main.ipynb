```python
import os
from dotenv import load_dotenv
load_dotenv()

import openai
openai.api_key = os.getenv("OPENAI_API_KEY")
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
model= ChatGroq(model="llama-3.3-70b-versatile", groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import SystemMessage, HumanMessage
messages=[
    SystemMessage(content="Translate the following from English to french"),
    HumanMessage(content="Hello, how are you?")
]

result = model.invoke(messages)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(result)
```

```python
## here we have 2 component on langchain 1. the model component which is the wrapper around the llm and 2. the output parser which is used to parse the output of the model and return it in a desired format. In this case we are using the StrOutputParser which will return the output as a string.
chain = model | parser
chain.invoke(messages)
```

```python
### prompt template
from langchain_core.prompts import ChatPromptTemplate

generic_template = "Translate the following into {LANGUAGE}:"

prompt = ChatPromptTemplate.from_messages(
    [("system", generic_template), ("user", "{TEXT}")]
)
```

```python
result = prompt.invoke({"LANGUAGE": "Urdu", "TEXT": "Hello, how are you?"})
```

```python
result.to_messages()
```

```python
### Chaining the prompt, model and parser together with LCEL 
chain = prompt | model | parser
chain.invoke({"LANGUAGE": "Urdu", "TEXT": "Hello, how are you?"})
```
